```python
#'company_info', 'api_data', 'other_details', 'topbox_information', 'jd_information',
def get_Frame(file):
    from pandas.io.json import json_normalize
    with open(file) as f:
        d = json.load(f)
    company_info = json_normalize(d['company_info'])
    df_company_info=pd.DataFrame(company_info)
    api_data = json_normalize(d['api_data'])
    df_api_data=pd.DataFrame(api_data)
    other_details = json_normalize(d['other_details'])
    df_other_details=pd.DataFrame(other_details)
    topbox_information = json_normalize(d['topbox_information'])
    df_topbox_information=pd.DataFrame(topbox_information)
    jd_information = json_normalize(d['jd_information'])
    df_jd_information=pd.DataFrame(jd_information)
    df=pd.concat([df_company_info, df_api_data,df_other_details,df_topbox_information,df_jd_information], axis=1)
    return df




```

```python
frames = [ get_Frame(f) for f in file_list ]
result = pd.concat(frames)
result['jd_id']=pd.to_numeric(result['jd_id'])
result.set_index('jd_id',inplace=True)
```

```python


def decontracted(phrase):
    # specific
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
```

```python

preprocessed_descreption = []

for sentance in Final['description'].values:
    sentance = re.sub(r"http\S+", "", sentance)
    sentance = BeautifulSoup(sentance, 'lxml').get_text()
    sentance = decontracted(sentance)
    sentance = re.sub("\S*\d\S*", "", sentance).strip()
    sentance = re.sub('[^A-Za-z]+', ' ', sentance)
    # https://gist.github.com/sebleier/554280
    sentance = ' '.join(e.lower() for e in sentance.split() if e.lower() not in stopwords.words())
    preprocessed_descreption.append(sentance.strip())
```
